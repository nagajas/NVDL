```python
import json

# Load the file (assuming it's in JSON format)
with open('/content/training_data_Subtask_2_train.json', 'r') as file:
    data = json.load(file)

# Extract text from conversations
conversations = []
for conv in data:
    for utterance in conv['conversation']:
        conversation_text = " ".join([utterance['text'] ])
        conversations.append(conversation_text)

# Assuming 'conversations' is a list of conversation text from previous steps
for i, conv in enumerate(conversations[:10]):
    print(f"Conversation {i + 1}:")
    print(conv)
    print("\n" + "-"*50 + "\n")  # Separator between conversations
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Tokenize the text
tokenized_conversations = [word_tokenize(conv) for conv in conversations]
print(tokenized_conversations)
```

```python
import gensim.downloader as api

# Load pre-trained GloVe embeddings
glove_vectors = api.load("glove-wiki-gigaword-300")

# Convert each token to its GloVe vector
def get_glove_embedding(word):
    try:
        return glove_vectors[word]
    except KeyError:
        # If the word is not in GloVe, return a zero vector
        return [0] * 300

# Apply GloVe embeddings to tokenized text
embedded_conversations = [[get_glove_embedding(word) for word in tokens] for tokens in tokenized_conversations]
```

```python
import numpy as np

ridx = np.random.randint(0, len(conversations))
print(f"Conversation {ridx+1}:")
w = tokenized_conversations[ridx][-2]
v = embedded_conversations[ridx][-2]
print(f"Vector size: {len(v)} d")
print(f"Word: {w}, GloVe Vector: {v}")
```

```python
for i, tokens in enumerate(tokenized_conversations[:10]):
    print(f"Conversation {i + 1}:")
    for word, vector in zip(tokens, embedded_conversations[i]):
        # Print only the first 5 dimensions of the GloVe vector for brevity
        print(f"Word: {word}, GloVe Vector (first 5 dimensions): {vector[:5]}")
    print()  # Print a new line for better readability
```
